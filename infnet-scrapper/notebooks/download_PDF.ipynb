# Downloading `pdf`

```python
import os, urllib2
import pandas as pd
import time
```

```python
SAVEDIR = '../data/PDF/'
```

```python
publications = pd.read_csv('../data/publications.csv')
```

```python
publications.head(3)
```

```python
doi_urls = publications.doi_url
pdf_urls = publications.pdf_url
pub_ids = publications.pub_id
```

```python
def download_pdf(pdf_url, doi_url, pub_id):
    saveFile = SAVEDIR + str(pub_id) + '.pdf'
    pdf_url = pdf_url.strip()
    if pdf_url == "UNKNOWN":
        print pub_id
        return
    assert pdf_url[-4:] == ".pdf"
    
    with open(saveFile, 'wb') as fout:
        try:
            response = urllib2.urlopen(pdf_url)
            fout.write(response.read())
        except:
            print(pub_id, pdf_url)
```

```python
for pdf_url, doi_url, pub_id in zip(pdf_urls, doi_urls, pub_ids):
    download_pdf(pdf_url, doi_url, pub_id)
    time.sleep(1)
```
